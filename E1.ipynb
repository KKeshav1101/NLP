```python
def ttest(w1, w2, c1, c2, c12):
    p1 = c1/N
    p2 = c2/N
    p12 = c12/N
    if p12==0:
        print("Not collocation")
        return
    obs = p12
    exp = p1*p2/N
    s2 = obs
    t = (obs-exp)/math.sqrt(s2/N)
    if t<CV:
        print("Not collocation")
    else:
        print("Collocation")
```

```python
def chi2test(w1, w2, c1, c2, c12):
    o11 = c12
    o12 = c1 - c12
    o21 = c2 - c12
    o22 = N - (o11+o12+o21)
    e11 = c1*c2/N
    e12 = c1*(N-c2)/N
    e21 = c2 * (N-c1)/N
    e22 = (N-c2)*(N-c1)/N
    chi2 = ((o11-e11)**2/e11)+((o12-e12)**2/e12)+((o21-e21)**2/e21)+((o22-e22)**2/e22)
    if chi2<CV:
        print("Not collocation")
    else:
        print("collocation")
```
