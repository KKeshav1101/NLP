```python
df = pd.read_excel('Bank.xlsx')
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
def df_convert(sense):
    return df[df['Sense of the word "Bank"']==sense]
```

```python
fin = df_convert('Financial Institution')
fin
```

```python
riv = df_convert('River Border')
riv
```

```python
test = df_convert('?')
```

```python
tot = len(fin)+len(riv)
```

```python
words= set()
for s in fin['Sentences for Training ']:
    words.update(s.split())
for s in riv['Sentences for Training ']:
    words.update(s.split())
```

```python
v = len(words)
```

```python
p_f = len(fin)/tot
p_r = len(riv)/tot
p_f,p_r
```

```python
def probability(df, w):
    c=0
    for i in df['Sentences for Training ']:
        if w in i:
            c+=1
    return (c+1)/(len(df)+v)
```

```python
def naive_bayes(df, prob, sent):
    prob = math.log2(prob)
    for w in sent:
        w_p = math.log2(probability(df,w))
        prob+=w_p
    return prob
```

```python
test
```

```python
for i in test['Sentences for Training ']:
    prob_f = naive_bayes(fin, p_f, i.split())
    prob_r = naive_bayes(riv, p_r, i.split())
    if prob_f>prob_r:
        print(i,'Finance')
    else:
        print(i,'River')
```
