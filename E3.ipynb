```python
import pandas as pd
import nltk
import math
```

```python
df = pd.read_csv('Attachment ambiguity WITH.csv', encoding='latin-1', header=None)
```

```python
corpus = []
for i in df[0]:
    corpus.extend(nltk.word_tokenize(i.lower()))
corpus = [i for i in corpus if i not in ['a','an','the']]
```

```python
count = {}
for w in corpus:
    count[w] = count.get(w,0)+1
count
```

```python
from nltk import bigrams
bigrams = list(bigrams(corpus))
```

```python
bigram_count = {}
for b in bigrams:
    bigram_count[b] = bigram_count.get(b,0)+1
bigram_count
```

```python
noun = input("Enter noun: ")
verb = input("Enter verb: ")
prep = input("Enter prep: ")
n = count.get(noun, 0)
v = count.get(verb, 0)
p_n = bigram_count.get((prep, noun), 0)
p_v = bigram_count.get((prep, verb), 0)
```

```python
def probability(n, v, p_n, p_v):
    prob_n = (p_n+1)/(n+len(count)+1)
    prob_v = (p_v+1)/(v+len(count)+1)
    return prob_n , prob_v
```

```python
def calculate_lambda(prob_n,prob_v):
    if prob_n==0:
        return float('inf')
    return math.log2(prob_v*(1-prob_n)/prob_n)
```

```python
prob_n, prob_v = probability(n, v, p_n, p_v)
lam = calculate_lambda(prob_n, prob_v)
if lam>0:
    print("Verb")
elif lam<0:
    print("Noun")
else:
    print("Not det")
```
